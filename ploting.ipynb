```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv("gas_prices.csv")
gas.head()
```

```python
plt.plot(gas.Year,gas.USA,label="USA")
plt.title("Gas prices Over Time (in USA)")
plt.plot(gas.Year,gas.UK,label="Uk")
plt.legend()
plt.show()
```

```python
for country in gas:
    if country !="Year":
        plt.plot(gas.Year,gas[country],marker= ".")
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("Us Dollers")
plt.legend()
plt.show()
```

```python
fifa=pd.read_csv("fifa_data.csv")
fifa.head()
```

```python
bins=[40,50,60,70,80,90]
plt.hist(fifa.Overall,bins=bins)
plt.ylabel("Number of Players")
plt.xlabel("Skill level")
plt.title("Distribution of player skills in FIFA 2018")
```

```python
import matplotlib.pyplot as plt
left=fifa.loc[fifa["Preferred Foot"]=="Left"].count()[0]
rigth=fifa.loc[fifa["Preferred Foot"]=="Rigth"].count()[0]
left,rigth
labels=["Left","Rigth"]
plt.pie([left,rigth],labels=labels)
plt.show()
```

```python
plt.style.use("default")
barcelona=fifa.loc[fifa.Club=="FC Barcelona"]["Overall"]
madrid=fifa.loc[fifa.Club=="Real Madrid"]["Overall"]
revs=fifa.loc[fifa.Club=="New England Revolution"]["Overall"]
labels=["FC Barcelona","Real Madrid","New England Revolution"]
plt.title("Professional Soccer Team Comparison")
plt.ylabel("FiFa overall rating")
plt.xlabel("FiFa overall rating")
boxes=plt.boxplot([barcelona,madrid,revs],labels=labels)
for box in boxes["boxes"]:
    box.set(color="#4286f4",linewidth=2)
plt.show()
```
